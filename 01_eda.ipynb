# Santander Transaction Prediction — Exploratory Data Analysis

```python
# Download data first — see data/README.md
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

train = pd.read_csv('../data/train.csv')
test  = pd.read_csv('../data/test.csv')
print('Train:', train.shape, '  Test:', test.shape)
```

```python
# Class distribution (~10:1 imbalance)
print(train['target'].value_counts())
train['target'].value_counts().plot(kind='bar', title='Target distribution')
plt.tight_layout()
```

```python
# Missing values
print('Missing values in train:', train.isnull().sum().sum())
print('Missing values in test: ', test.isnull().sum().sum())
```

```python
# Distribution of a few features by target
features = [c for c in train.columns if 'var_' in c][:6]
fig, axes = plt.subplots(2, 3, figsize=(15, 8))
for ax, col in zip(axes.ravel(), features):
    train[train.target == 0][col].hist(ax=ax, bins=50, alpha=0.6, label='0')
    train[train.target == 1][col].hist(ax=ax, bins=50, alpha=0.6, label='1')
    ax.set_title(col)
    ax.legend()
plt.tight_layout()
```
